```python
# Cargamos el archvivo 
# referencia 
#https://colab.research.google.com/github/computationalcore/introduction-to-python/blob/master/notebooks/4-files/PY0101EN-4-1-ReadFile.ipynb#scrollTo=ArrqzEmQx2_M

base = "/content/drive/MyDrive/CodingDojo/DataScience/Python_DataScience/Ejercicios/SenseAndSensibilityCh1.txt"

archivo = open(base, "r")
archivoContenido = archivo.read()
#print(archivoContenido)
```

```python
# Se limpia la cadena

limpiar = ",.\/:;"

for caracter in limpiar:
  archivoContenido = archivoContenido.replace(caracter,"")    

# se transforma a minusculas
archivoContenido = archivoContenido.lower()

# separamos por espacios
palabras = archivoContenido.split(" ")
  
dictFrecuencia = {}

for palabra in palabras:
  if palabra in dictFrecuencia:
    dictFrecuencia[palabra] +=1
  else:
    dictFrecuencia[palabra] =1
 
 

for palabra in dictFrecuencia:
  frecuencia =  dictFrecuencia[palabra]
  print(f"{palabra}:{frecuencia}")

```
